```python
#prediction starting
# selecting the features V95-V137 && V279-321
features = ['V%d' % number for number in range(95,138)]
featuesAppend = ['V%d' % number for number in range(279,322)]
features.extend(featuesAppend)
#select Target variable which should be predicted
target = 'isFraud'
#create training data X and Y
X_data = df_train_transaction[features]
y_data = df_train_transaction[target]
```

##### remove NAN in features

```python
# remove all NAN from traniing data, replace by 0 
X_data = X_data.fillna(0)
#check of succefull replaced
print("Number of NAN in training features :" )
X_data.isna().sum().iloc[::10]
```

##### normalize function from examle site

```python
def normalize(X):
    """
    Make the distribution of the values of each variable similar by subtracting the mean and by dividing by the standard deviation.
    """
    for feature in X.columns:
        X[feature] -= X[feature].mean()
        X[feature] /= X[feature].std()
    return X
```

#### initalize model, datasplitter, report 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.metrics import classification_report
#model for prediction
LRmodel = LogisticRegression()

```

##### do the data splitting && training && fitting && reporting 

```python
#define the splitter for splitting the data in train and test set
splitter = StratifiedShuffleSplit(n_splits=1, test_size=0.5, random_state=0)

#loop trough splits
for train_indx, test_indx in splitter.split(X_data,y_data):
    X_train, y_train = X_data.iloc[train_indx],y_data.iloc[train_indx]
    X_test, y_test = X_data.iloc[test_indx], y_data.iloc[test_indx]
    
    #normalize Test Data
    X_train = normalize(X_train)
    X_test = normalize(X_test)


    #fit the model
    LRmodel.fit(X_train, y_train)
    #predict 
    y_pred = LRmodel.predict(X_test)

    #get the data for the prediction
    print(classification_report(y_test,y_pred))
```
